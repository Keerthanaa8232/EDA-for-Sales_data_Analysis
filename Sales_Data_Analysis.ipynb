WHICH PRODUCT SOLD THE MOST AND WHY?
---> user ratings
---> Low price

```python
all_data.columns
```

```python
count_df = all_data.groupby (['Product']).agg({'Quantity Ordered':'sum', 'Price Each': 'mean'})
count_df
```

```python
count_df = count_df.reset_index()
count_df
```

```python
fig, ax1 = plt.subplots(figsize=(12,6))
ax2 = ax1.twinx()

# Bar on ax1
ax1.bar(count_df['Product'], count_df['Quantity Ordered'], color= 'g')

# Line on ax2
ax2.plot(count_df['Product'], count_df['Price Each'], color='b', marker='o')
ax1.set_ylabel('Quantity Ordered')
ax2.set_ylabel(' Avg Price ($)')

ax1.set_xticklabels(count_df['Product'], rotation=90)
plt.show()
```

PRODUCTS THAT ARE OFTEN SOLD TOGETHER

```python
all_data.columns
```

```python
all_data['Order ID']
```

```python
all_data['Order ID'].duplicated(keep= False)
```

```python
df_duplicated = all_data[all_data['Order ID'].duplicated(keep= False)]
df_duplicated
```

```python
df_duplicated.groupby(['Order ID'])['Product'].apply(lambda x: ','.join(x))
```

```python
df_duplicated.groupby(['Order ID'])['Product'].apply(lambda x: ','.join(x)).reset_index()
```

```python
dup_products = df_duplicated.groupby(['Order ID'])['Product'].apply(lambda x: ','.join(x)).reset_index().rename(columns={'Product':'grouped_product'})
```

```python
dup_products
```

```python
 dup_products_df = df_duplicated.merge(dup_products, how= 'left',on = 'Order ID')
```

```python
dup_products_df
```

```python
no_dup_df = dup_products_df.drop_duplicates(subset=['Order ID'])
no_dup_df
```

```python
no_dup_df.shape
```

```python
no_dup_df['grouped_product'].value_counts()
```
